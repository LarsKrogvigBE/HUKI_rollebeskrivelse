```python
import re
import string
import csv
import docx
import numpy as np

from docx import Document
```

```python
# INPUT

cell_sht1_top_left     = 'E7'    # Øverste venstre celle av HUKI-matrisen (under kolonnenavnene og til høyre for aktivitetbeskrivelse)
cell_sht1_bottom_right = 'BG866' # Nederste høyre celle av HUKI-matrisen

cell_sht2_top_left     = 'B5'    # Øverste venstre celle av av arket Formål og overordnede (under kolonnenavnene og til høyre for cellen "Rollens hensikt")
cell_sht2_bottom_right = 'BD7'   # Nederste høyre celle av av arket Formål og overordnede
```

```python
# TEKST

ansvar = ['H','U','K','I']
ansvartext = {'H': 'Rollen er hovedansvarlig for at følgende aktiviteter blir utført, og for å ta beslutninger knyttet til disse aktivitetene',
              'U': 'Rollen skal utføre følgende aktiviteter',
              'K': 'Rollen skal være tilgjengelig for konsultasjon i følgende aktiviteter',
              'I': 'Rollen skal informeres om følgende aktiviteter'} 
```

```python
## Lesing av filer

rowset = []
with open('HUKI for prosesser i Feiring Bruk v2.csv', 'rb') as csvfile:
    csvfile.seek(0)
    reader = csv.reader(csvfile, delimiter=';')
    
    rowset = []
    i = 0
    for row in reader:
        rowset.append(row)
        
D = np.array(rowset,dtype=np.string_)

with open('HUKI for prosesser i Feiring Bruk v2 beskrivelse.csv', 'rb') as csvfile:
    reader = csv.reader(csvfile, delimiter=';')
    rowset = []
    for row in reader:
        rowset.append(row)

D1 = np.array(rowset,dtype=np.string_)
```

```python
## Uthenting av koordinater

def cell2idx(cell):
    ord0 = ord('A') - 1
    numchar = ord('Z') - ord0
    match = re.search('([A-Z]+)([1-9][0-9]*)',cell)
    
    if match:
        row = int(match.group(2)) - 1
        col = -1
        for j, char in enumerate(match.group(1)[::-1]):
            col += (ord(char) - ord0) * pow(numchar, j)
    else:
        col = None
        row = None
        print "ERROR - Invalid cell"
        
    return row, col


(i0, j0) = cell2idx(cell_sht1_top_left)
(i1, j1) = cell2idx(cell_sht1_bottom_right)

(i2, j2) = cell2idx(cell_sht2_top_left)
(i3, j3) = cell2idx(cell_sht2_bottom_right)

# Koordinater
row_avd        = i0 - 4
row_avd_rolle  = i0 - 3
row_rolle      = i0 - 2
row_data_start = i0

row_rolle_hen  = i2 - 2
row_hensikt    = i2
row_overord    = i2 + 1
row_underord   = i2 + 2
row_standard   = i2 + 7

col_hovedaktivitet = j0 - 3
col_underaktivitet = j0 - 1
col_data_start     = j0
col_data_end       = j1

col_beskr_start    = j2 
col_beskr_end      = j3
col_standard       = j2 - 1 

avd       = set(D[row_avd, col_data_start:col_data_end+1])
rolle     = D[row_rolle, col_data_start:col_data_end+1]
avd_rolle = D[row_avd_rolle, col_data_start:col_data_end+1]
hdesc     = D[row_data_start:, col_hovedaktivitet]
desc      = D[row_data_start:, col_underaktivitet]
data      = D[row_data_start:, col_data_start:col_data_end+1]

hensikt   = D1[row_hensikt, col_beskr_start:col_beskr_end+1]
overord   = D1[row_overord, col_beskr_start:col_beskr_end+1]
underord  = D1[row_underord, col_beskr_start:col_beskr_end+1]
standard  = D1[row_standard, col_standard]

rolle_postfix = [ ' ('+avd_rolle[j]+')' if len([rll for rll in rolle if rll==rl]) > 1 else '' for j, rl in enumerate(rolle)]

(m,n) = data.shape
```

```python
def hent_dokumentliste(filnavn):
    f = open(filnavn, 'r')
    lines = f.read().splitlines()
    f.close()
    
    lines = [string.split(line, ':') for line in lines]
    dokumentliste = dict(zip([t[0] for t in lines], [[s.strip() for s in string.split(t[1], ',')] for t in lines]))
    
    return dokumentliste

def stringprep(s):
    return unicode(s.decode('utf-8')) # document must be saved in utf-8..

def stringclean(s):
    tmp = re.search('([a-zæøåA-zÆØÅ0-9 _]*)',s)
    return unicode(tmp.group(0).decode('utf-8'))

def hent_dokumentdata():
    # Denne funksjonen må vite om rolle, data, desc, og så videre
    dokumentdata = {}
    for j, rl in enumerate(rolle):
        doc = {'idx': j,
               'rolle': rl,
               'rolle_postfix': rolle_postfix[j],
               'avdeling': avd_rolle[j],
               'hensikt': hensikt[j],
               'overordnet': overord[j],
               'underordnet': underord[j],
               'ansvar': {}}
        for a in ansvar:
            idx_list  = [i for i in xrange(m) if a in data[i,j]] 
            ansv_list = [{'desc': desc[i]} if not desc[i].strip()=='' else {'desc': hdesc[i]} for i in idx_list]
            
            dlist = dict(zip(idx_list, ansv_list))
            
            for i in dlist:
                
                if sum([1 for k in xrange(n) if 'U' in data[i,k]]) > 1:
                    dlist[i]['konsultere'] = [rolle[k] + rolle_postfix[k] for k in xrange(n) if 'K' in data[i,k] and avd_rolle[k] == avd_rolle[j]] \
                                             + [rolle[k] + rolle_postfix[k] for k in xrange(n) if 'K*' in data[i,k] and not avd_rolle[k] == avd_rolle[j]]
                    dlist[i]['informere']  = [rolle[k] + rolle_postfix[k] for k in xrange(n) if 'I' in data[i,k] and avd_rolle[k] == avd_rolle[j]] \
                                             + [rolle[k] + rolle_postfix[k] for k in xrange(n) if 'I*' in data[i,k] and not avd_rolle[k] == avd_rolle[j]]
                else:
                    dlist[i]['konsultere'] = [rolle[k] + rolle_postfix[k] for k in xrange(n) if 'K' in data[i,k]]
                    dlist[i]['informere']  = [rolle[k] + rolle_postfix[k] for k in xrange(n) if 'I' in data[i,k]]    
            if dlist: doc['ansvar'][a] = dlist
        
        dokumentdata[(avd_rolle[j], rl)] = doc
        
    return dokumentdata

def generate_document(dokumentnavn, inneholdsliste, dokumentdata):
    
    ##################################################################################
    def insert_document_part(doc):
    
        # Tekst om rollens hensikt
        document.add_paragraph(u'Rollens hensikt', style='Heading 3')
        document.add_paragraph(stringprep(doc['hensikt']))
        
        # Tabell med rapportering
        document.add_paragraph(u'Rapportering', style='Heading 3')
        
        table = document.add_table(rows=1, cols=2, style='Table Grid BE-style')
        hdr = table.rows[0].cells
        hdr[0].text = stringprep('Nærmeste overordnede til rollen:')
        hdr[1].text = stringprep('Rollen er nærmeste overordnede til følgende roller:')
        row = table.add_row().cells
        row[0].text = stringprep(doc['overordnet'])
        row[1].text = stringprep(doc['underordnet'])
        
        # HUKI
        document.add_paragraph(u'Ansvar og myndighet', style='Heading 3')
        
        for a in ansvar:
            
            document.add_paragraph(stringprep(ansvartext[a]), style='Heading 4')
            
            # List opp ansvar for a = 'H', 'U', 'K' eller 'I'
            if a in doc['ansvar']:
                
                # Spesiell behandling dersom rollen har utførende ansvar (a = 'U')
                if a=='U':
                    
                    table = document.add_table(rows=1, cols=3, style='Table Grid BE-style')
                    
                    hdr = table.rows[0].cells
                    hdr[0].text = stringprep('Skal utføre følgende aktiviteter:')
                    hdr[1].text = stringprep('Skal konsultere følgende roller:')
                    hdr[2].text = stringprep('Skal informere følgende roller:')
                    for i in sorted(doc['ansvar']['U'].keys()):
                        d = doc['ansvar']['U'][i]
                        row = table.add_row().cells
                        row[0].text = stringprep(d['desc'])
                        row[1].text = stringprep(', '.join(d['konsultere']))
                        row[2].text = stringprep(', '.join(d['informere']))
                    
                    document.add_paragraph(u'\n'+stringprep(standard))
                    
                else:
                    for i in sorted(doc['ansvar'][a].keys()):
                        d = doc['ansvar'][a][i]
                        document.add_paragraph(stringprep(d['desc']), style='Bullet 1')
            else:
                document.add_paragraph(stringprep('Ingen aktiviteter'))
        
        pass
    ##################################################################################
    
    document = Document('blank.docx')
    
    print 'Genererer Rollebeskrivelse for', dokumentnavn
    
    # Alle roller "utenfor" aveldeling, som f.eks Produksjonsansvarlig
    for rl in [r for r in inneholdsliste if r in rolle]:
        
        document.add_paragraph(stringprep(rl), style='Heading 1')
        
        doc = dokumentdata[('Overordnet', rl)]
        insert_document_part(doc)
    
    # Alle avdelinger i inneholdslisten
    for avd in [a for a in inneholdsliste if a not in rolle]:
        
        document.add_paragraph(stringprep(avd), style='Heading 1')
        
        # Alle roller for disse avdelingene
        for rl in [r for j, r in enumerate(rolle) if avd_rolle[j]==avd]:
            
            document.add_paragraph(stringprep(rl), style='Heading 2')
            
            doc = dokumentdata[(avd, rl)]
            insert_document_part(doc)
        
            
    document.save(u'Genererte dokumenter/'+stringclean(dokumentnavn)+u'.docx')
    
    pass  
```

```python
dokumentliste = hent_dokumentliste('dokumentliste.txt')
dokumentdata = hent_dokumentdata()

for d in dokumentliste:
    generate_document(d, dokumentliste[d], dokumentdata)
```
